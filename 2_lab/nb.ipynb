> програми можуть писатись як набір команд які потрібно виконати для отримання результату
> або ми можемо використати команди мови програмування для опису певного фізичного явища
> у базовому представленні, числа є також абстракцією, наприклад ми можемо користуватись або арабськими або римськими цифрами, однак абстаркція полягає  у тому що цифри це є відображення кількості чогось

Будь-що у Пайтон описується абстракціями, і вони співвідносяться з парою `клас` і `обєкт`
Клас - це загальне уявлення про абстрації, її опис, як саму абстрація повинна бути представлена, з якими характеристиками та які властивості повинна мати. Для прикладу, у грі є персонажі, ці персонажі можуть мати певний клас, мажуть мати набір склів, мажуть мати вигляд та певні ійтеми, Все що може з себе представляти персонаж це є КЛАС, томущо клас описує що таке персонаж в грі. Клас це абстрація як описує цифрового аватара у грі.
Обєкт - це конкретне вираження одного унікального представлення класу, або як в книжках пишуть - екземпляр класу. Для поснення з грою, Ваш ігровий персонаж є обєктом класу Персонаж, і має конкретний клас/расу Воїн/Ельф, має конкретний набір скілів, вигляд та якісь ітеми.

Приклад: В грі є ітеми, наприклад мечі. Меч це є абстракція, тому він є класом `МЕЧ`. Ви вибили певний ітем меча, `короткий меч` який має вигляд катани, з ручкою синього кольору, якість зелена (продвинута), та має характеристику атаки +15 і додає силу +2, тому цей меч є обєктом класу `МЕЧ`.

Поки ми дивились лише на вбудовані класи та обєкти які програмісти створювали до нас. Однак ми також можемо описувати певні явища і створювати класти та обєкти. Далі давайте створимо власне наш перший клас.

Ще клас можна уявити собі як шаблон а відповідно обєкт це кокретна "річ" зроблена по шаблону.
Ми спробуємо описати у програмі Мечів та процес отримання конкретного меча гарвцем.

```python
swords_name = ["короткий меч", "довгий меч", "катана", "рапіра"]
swords_rarity = ["Green", "Blue", "Exotic", "Epic"]
swords_damag = [2, 5, 3, 5]
swords_vitality = [10, 10, 15, 20]

from random import randint
print(f"моделюємо випадання якогось числа, що буде вказувати на меч")
i = randint(0, len(swords_name)-1)
print(f"Вітаємо, Вам випав меч! Стати:\nНазва: {swords_name[i]}\nРідкість: {swords_rarity[i]}\nДамаг: +{swords_damag[i]}\nВитривалість: +{swords_vitality[i]}")
```

Давайте спробуємо зробити такий самий меч тільки з використанням ООП, у самий найпростіший спосіб з використанням динмічних атрибутів

```python
class Swords:
    pass

# створюємо перший обєкт класу Мечі
s1 = Swords()
# Дигнамічно задаємо атрибути
s1.name = "короткий меч"
s1.rarity = "Green"
s1.damag = 2
s1.vitality = 10
s1.bonus = "Немає"

s2 = Swords()
s2.name = "довгий меч"
s2.rarity = "Blue"
s2.damag = 5
s2.vitality = 10
s2.bonus = "Спритність"

s3 = Swords()
s3.name = "катана"
s3.rarity = "Exotic"
s3.damag = 3
s3.vitality = 15
s3.bonus = "Берсерк"

s4 = Swords()
s4.name = "рапіра"
s4.rarity = "Epic"
s4.damag = 5
s4.vitality = 20
s4.bonus = "Швидкі удари"

swords = [s1, s2, s3, s4]
i = randint(0, len(swords)-1)
s = swords[i]
print(f"Вітаємо, Вам випав меч! Стати:\nНазва: {s.name}\nРідкість: {s.rarity}\nДамаг: +{s.damag}\nВитривалість: +{s.vitality}\nУнікальна характеристика: {s.bonus}")
```

У попередньому прикладі, клас був пустим і всі його аргументи задавались динамічно. В такому варіанті створення обєктів не гарантується що обєкти будуть однакові, та містити в собі спільні поля.

Дану інформацію видає ChatGPT розповідаючи що таке методи __init__ та __new__ .

> У Пайтоні конструктор класу - це спеціальний метод з ім'ям __init__. Цей метод викликається автоматично при створенні нового об'єкта цього класу. Він призначений для ініціалізації атрибутів об'єкта.

> Метод __new__ в Python є статичним методом класу, який викликається для створення нового екземпляра класу перед тим, як буде викликаний метод __init__. Його основна функція - це створення нового об'єкта. Він приймає клас (як аргумент cls) та будь-які інші аргументи, які передаються підчас створення екземпляру.

Тому, всі атрибути/поля потрібно передбачати в класі який виступає шаблоном обєкта

```python
class Swords:
    who_has_buff = [] # Ця класова змінна відслідковує на які обєкти зараз накладено баф
    
    def __init__(self, name:str, rarity:str, damag:int, vitality:int, bonus:str = None) -> None: # Це є свого роду конструктор
        """Конструктор для створення обєкту Меч.
        name: поля для імені;
        rarity: Рідкість предмету; 
        damag: ;
        vitality: ;
        bonus: необовязковий аргумент;
        """
        self.name = name # це є атрибути обєкта, їх можна змінювати після створення обєкту
        self.rarity = rarity
        self.damag = damag
        self.vitality = vitality
        self.bonus = bonus

        self.buff_damage = 0
        self.buff_vitality = 0
    
    def get_buff_damag(self, damag:int) -> str:
        if self.__hash__ in Swords.who_has_buff:
            return f"На меч {self.name} вже накладено баф"
        self.buff_damage = damag
        Swords.who_has_buff.append(self.__hash__)
        return f"Накладено баф на {self.name} який додає атрибут нанесення шкоди +{damag}"
    
    def get_buff_vitality(self, vitality:int) -> str:
        if self.__hash__ in Swords.who_has_buff:
            return f"На меч {self.name} вже накладено баф"
        self.vitality += vitality
        self.buff_vitality = vitality
        Swords.who_has_buff.append(self.__hash__)
        return f"Накладено баф на {self.name} який додає атрибут здоровя +{vitality}, загальне здоровя: {self.vitality}"
    
    def expired_buff(self) -> str:
        Swords.who_has_buff.remove(self.__hash__)
        if self.buff_damage > 0:
            self.buff_damage = 0
            return "Дія бафу на нанасення шкоди завершилась!"
        if self.buff_vitality > 0:
            self.vitality -= self.buff_vitality
            self.buff_vitality = 0
            return "Дія бафу на здоровя завершилась!"
        return "На мечі не має ніякого бафу!"
    
    def attack(self, item:Swords = None) -> str:
        """Метод для атаки
        """
        if isinstance(item, Swords):
            item.vitality -= self.hit
            return f"Нанесено шкоду {self.hit} мечу {item.name}"
        else:
            return "Ми промахнулись!"
        
    def parry(self, damage) -> int:
        """Метод для парирування, тут в нас є бага, ми не можемо зменшувати проперті health
        """
        self.vitality -= damage
        return self.vitality
    
    @property 
    def get_name(self) -> str:
        """Повертає значення імені. І тут проперті змінювати не можна! Це моле лише для читання."""
        return self.name
    
    @property
    def info(self) -> str:
        """Проперті для читання, виводить інформацію про обєкт."""
        return f"""\n<<<<<Стати:
        Назва: {self.get_name}
        Рідкість: {self.rarity}
        Дамаг: +{self.hit}
        Витривалість: +{self.health}
        Унікальна характеристика: {self.bonus}>>>>>\n"""

    @property
    def hit(self):
        """Проперті яка визначає загальне значення нанесення шкоди"""
        return self.damag + self.buff_damage
    
    @property
    def health(self):
        """Проперті яка визначає всю витривалість ХП який має обєкт"""
        return self.vitality

a = Swords("Короткий меч", "Green", 2, 10)
b = Swords("Довгий меч", "Blue", 5, 10)
# b = Swords.__init__() # тут виникне помилка, бо ми не передаємо вказівник на обєкт
# print(a.__init__()) # це був приклад для пояснення self, зараз він не працює

for item in [a, b]:
    print(type(item), f"Створено обєкт {item}\nВ якого є атрибути {item.__dict__}")
```

Створений клас є класичною основою створення класів у Пайтон, ми використали метод `__init__` та прописали всі атрибути якими повинен володіти обєкт нашого класу.
Для взаємодії або якщо ми хочемо здійснювати певні дії над мечем або власне викликати дії самого меча, ми повинні створювати так звані методи.
Створимо метод який накладає баф на наш меч.
Створимо методи який буде атакувати з нашим мечем.

```python
a = Swords("Короткий меч", "Green", 2, 10)
b = Swords("Довгий меч", "Blue", 5, 10)
print(f"Початкове значення шкоди меча {a.name} має значення {a.damag}")
print(a.get_buff_damag(2))
print(f"Після накладення бафу на меч {a.name} шкода буде мати значення {a.damag} та витривалість меча {a.vitality}")
print(f"Після того як {b.name} з шкодою {b.damag} наніс удар {a.name} в нас залишилось витривалості {a.parry(b.damag)}")
print(f"Завдаємо удару у відповідь {a.name} по {b.name}, будемо мати {a.attack(b)}. Після атаки витривалість {b.name} буде {b.vitality}")
```
